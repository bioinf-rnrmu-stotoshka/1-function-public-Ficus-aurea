```python
'Fibonacci Numbers '
from numbers import Number
def g_f(n: int) -> Number:

    assert isinstance(n, int)
    assert 0 <= n <= 25

    if n==0:
        return 0
    if n==1:
        return 1
    if n>1:
        return g_f(n-1) + g_f(n-2)
e = 22
f = g_f(e)
print(f)
```

```python
'Degree Array'
from numbers import Number
from typing import List
def f_d(n: int, edges: List[List[int]]) -> List[int]:
    
    assert (isinstance(n, int) and n > 0)
    assert isinstance(edges, list)
    degrees = [0] * (n + 1)

    for edge in edges:
        assert (isinstance(edge, list) and len(edge) == 2)

        u, v = edge
        assert isinstance(u, int) and isinstance(v, int)
        assert (1 <= u <= n and 1 <= v <= n)

        degrees[u] += 1
        degrees[v] += 1
    return degrees[1:]
c = 6
e = [[1, 2], [2, 3], [6, 3], [5, 6], [2, 5], [2, 4], [4, 1]]
d = f_d(c, e)
print(d)
```

```python
'Insertion Sort'
from numbers import Number
from typing import List
def f_s(n: int, numbers: List[int]) -> Number:
    
    assert isinstance(n, int) and n > 0
    assert isinstance(numbers, list)
    assert len(numbers) == n

    s = 0
    for i in range(1, n):
        k = i
        while k >= 1 and numbers[k] < numbers[k - 1]:
            numbers[k], numbers[k - 1] = numbers[k - 1], numbers[k]
            k -= 1
            s += 1
        assert isinstance(numbers[k-1], (int, float, complex)) and not isinstance(numbers[k-1], bool)
    return s
c1 = 6
e = [9, 10, 3, 5, 1, 2]
c2 = f_s(c1, e)
print(c2)
```

```python
'Counting DNA Nucleotides'
from typing import Tuple
def f_c_dna(line: str) -> Tuple[int, int, int, int]:
   
    assert isinstance(line, str)
    assert (0 < len(line) <= 1000)

    count_a, count_c, count_g, count_t = 0, 0, 0, 0
    for i in line:
        assert i in 'ACGT'
        if i == "A":
            count_a += 1
        if i == "C":
            count_c += 1
        if i == "G":
            count_g += 1
        if i == "T":
            count_t += 1
    return count_a, count_c, count_g, count_t
s_dna = "AGCTTTTCATTCTGACTGCAACGGGCAATATGTCTCTGTGTGGATTAAAAAAAGAGTGTCTGATAGCAGC"
c_dna = f_c_dna(s_dna)
print(c_dna[0],c_dna[1],c_dna[2],c_dna[3])
```

```python
'Transcribing DNA into RNA'
def t(line: str) -> str:
    
    assert isinstance(line, str)
    assert (0 < len(line) <= 1000)

    c_dna = ""
    for i in line:
        assert i in "ACGT"
        if i == "T":
            c_dna += "U"
        else:
            c_dna += i
    return c_dna
s_dna = "GATGGAACTTGACTACGTAAATT"
s_rna = t(s_dna)
print(s_rna)
```

```python
'Complementing a Strand of DNA'
def r(line: str) -> str:
    
    assert isinstance(line, str)
    assert (0 < len(line) <= 1000)

    n_dna = ""
    for i in line:
        assert i in "ACGT"
        if i == "A":
            n_dna += "T"
        elif i == "C":
            n_dna += "G"
        elif i == "G":
            n_dna += "C"
        elif i == "T":
            n_dna += "A"
    return n_dna[::-1]
s_dna = "AAAACCCGGT"
r_dna = r(s_dna)
print(r_dna)
```

```python
'Rabbits and Recurrence Relations'
from numbers import Number
def b_r(n: int, k: int) -> Number:
    
    assert isinstance(n, int) and isinstance(k, int)
    assert n <= 40
    assert k <= 5

    if n == 1:
        return 1
    if n == 2:
        return 1
    return b_r(n - 1,k) + b_r(n - 2,k) * k
d = 5
o = 3
a_r = b_r(d, o)
print(a_r)
```
